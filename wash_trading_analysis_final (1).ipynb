```python
def detect_nft_cycles(df, time_window_days=30, max_cycle_length=4):
    # Group transactions by NFT_ID
    nft_groups = df.groupby('NFT_ID')
    
    cycles = []
    for nft_id, group in nft_groups:
        G = nx.MultiDiGraph()
        # Add edges for this NFT
        for _, row in group.iterrows():
            G.add_edge(row['Seller_address'], row['Buyer_address'],
                       nft_id=nft_id, price=row['Price_USD'],
                       timestamp=row['Datetime_updated_seconds'])
        
        # Custom DFS to find cycles
        def dfs_cycle(node, start_node, path, visited, edges_used, max_timestamp):
            if len(path) > max_cycle_length:
                return
            if node == start_node and len(path) > 1:
                cycle_edges = []
                valid = True
                for i in range(len(path) - 1):
                    u, v = path[i], path[i + 1]
                    edge_data = G.get_edge_data(u, v, default={})
                    if not edge_data or max_timestamp - list(edge_data.values())[0]['timestamp'] > time_window_days * 86400:
                        valid = False
                        break
                    cycle_edges.append((u, v, list(edge_data.values())[0]))
                if valid:
                    cycles.append(cycle_edges)
                return
            
            for neighbor in G.neighbors(node):
                edge_key = (node, neighbor)
                if edge_key not in visited:
                    visited.add(edge_key)
                    path.append(neighbor)
                    dfs_cycle(neighbor, start_node, path, visited, edges_used, max_timestamp)
                    path.pop()
                    visited.remove(edge_key)
        
        max_ts = group['Datetime_updated_seconds'].max()
        for node in G.nodes():
            dfs_cycle(node, node, [node], set(), set(), max_ts)
    
    return cycles

# Apply to a subset for testing (e.g., first 1000 rows or specific collection)
subset = opensea.head(1000)  # Adjust as needed
cycles = detect_nft_cycles(subset)
print(f"Detected {len(cycles)} cycles in subset:")
for i, cycle in enumerate(cycles[:5]):  # Limit output
    print(f"Cycle {i + 1}: {cycle}")
```

```python
from collections import defaultdict
import networkx as nx  # Ensure this is also included

def detect_sccs(df, min_occurrences=50):
    suspicious_sccs = {}
    collection_groups = df.groupby('Smart_contract')
    
    for contract, group in collection_groups:
        G = nx.DiGraph()
        for _, row in group.iterrows():
            G.add_edge(row['Seller_address'], row['Buyer_address'])
        
        edge_counts = defaultdict(int)
        for u, v in G.edges():
            edge_counts[(u, v)] += 1
        
        G_simplified = nx.DiGraph()
        G_simplified.add_edges_from(edge_counts.keys())
        scc_counts = defaultdict(int)
        
        while G_simplified.edges():
            sccs = list(nx.strongly_connected_components(G_simplified))
            for scc in sccs:
                if len(scc) > 1:
                    scc_counts[frozenset(scc)] += 1
            edges_to_remove = []
            for u, v in G_simplified.edges():
                edge_counts[(u, v)] -= 1
                if edge_counts[(u, v)] == 0:
                    edges_to_remove.append((u, v))
            G_simplified.remove_edges_from(edges_to_remove)
        
        suspicious_sccs[contract] = {scc: count for scc, count in scc_counts.items() if count >= min_occurrences}
    
    return suspicious_sccs

# Now run the SCC detection
sccs = detect_sccs(subset_multi_full, min_occurrences=50)
print(f"Detected suspicious SCCs:")
for contract, scc_dict in sccs.items():
    print(f"Collection {contract}: {len(scc_dict)} SCCs")
    for scc, count in list(scc_dict.items())[:5]:
        print(f"  SCC {list(scc)}: {count} occurrences")
```

```python
def detect_rapid_sequences(df, time_window_hours=12, price_deviation=0.05):
    suspicious_sequences = []
    df_sorted = df.sort_values('Datetime_updated_seconds')
    
    for nft_id in df_sorted['NFT_ID'].unique():
        nft_trades = df_sorted[df_sorted['NFT_ID'] == nft_id]
        if len(nft_trades) < 2:
            continue
        
        for i in range(len(nft_trades) - 1):
            start_trade = nft_trades.iloc[i]
            end_trade = nft_trades.iloc[i + 1]
            time_diff = end_trade['Datetime_updated_seconds'] - start_trade['Datetime_updated_seconds']
            if time_diff <= time_window_hours * 3600:
                price_change = abs(end_trade['Price_USD'] - start_trade['Price_USD']) / start_trade['Price_USD']
                if price_change <= price_deviation:
                    suspicious_sequences.append((start_trade['Transaction_hash'], end_trade['Transaction_hash']))
    
    return suspicious_sequences
```

# 1. Temporal Clustering with Anomalous Price Dynamics
Idea: Wash traders often execute rapid trades to inflate volume, but their price patterns may deviate from organic market trends (e.g., stable prices despite low liquidity or sudden spikes with no external trigger). Cluster transactions temporally and flag clusters with anomalous price behavior.
Implementation:
Use DBSCAN or HDBSCAN to cluster transactions by Datetime_updated_seconds within small time windows (e.g., 1 hour).
For each cluster, compute price variance, trade frequency, and address overlap.
Flag clusters with:
High trade frequency (>10 trades/hour).
Low price variance (<5%) or extreme spikes (>100% change without market-wide correlation).
High address repetition (e.g., same Seller_address or Buyer_address in >50% of trades).

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Normalize timestamps for clustering
X = subset_multi_full['Datetime_updated_seconds'].values.reshape(-1, 1) / 3600  # Hours
db = DBSCAN(eps=1, min_samples=5).fit(X)  # 1-hour radius, min 5 trades
subset_multi_full['Cluster'] = db.labels_

suspicious_clusters = []
for cluster_id in subset_multi_full['Cluster'].unique():
    if cluster_id == -1: continue  # Noise points
    cluster = subset_multi_full[subset_multi_full['Cluster'] == cluster_id]
    price_var = cluster['Price_USD'].var()
    addr_overlap = len(cluster['Seller_address'].value_counts()) / len(cluster)
    if len(cluster) > 10 and (price_var < 0.05 or price_var > 100) and addr_overlap > 0.5:
        suspicious_clusters.append(cluster)
print(f"Detected {len(suspicious_clusters)} suspicious temporal clusters")
```

```python
import pandas as pd
import networkx as nx
from community import best_partition  # pip install python-louvain
import numpy as np

# Build the trading graph
G = nx.DiGraph()
for _, row in subset_multi_full.iterrows():
    G.add_edge(row['Seller_address'], row['Buyer_address'], weight=row['Price_USD'])

# Detect communities using Louvain algorithm
partition = best_partition(G.to_undirected())  # Undirected for community detection
communities = {}
for node, comm_id in partition.items():
    if comm_id not in communities:
        communities[comm_id] = []
    communities[comm_id].append(node)

# Analyze communities for suspicious activity
suspicious_communities = []
comm_stats = []

for comm_id, nodes in communities.items():
    if len(nodes) < 3: continue  # Skip small communities
    subgraph = G.subgraph(nodes)
    internal_trades = [(u, v) for u, v in subgraph.edges() if u in nodes and v in nodes]
    if not internal_trades: continue
    
    trade_df = subset_multi_full[
        (subset_multi_full['Seller_address'].isin(nodes)) & 
        (subset_multi_full['Buyer_address'].isin(nodes))
    ]
    volume = trade_df['Price_USD'].sum()
    trade_count = len(trade_df)
    avg_volume_per_trade = volume / trade_count if trade_count > 0 else 0
    
    # Flag communities with high internal volume or trade frequency
    if trade_count > 50 and (volume > 10000 or avg_volume_per_trade > 1000):
        suspicious_communities.append(trade_df)
        comm_stats.append({
            'Community_ID': comm_id,
            'Address_Count': len(nodes),
            'Trade_Count': trade_count,
            'Volume_USD': volume,
            'Avg_Volume_Per_Trade': avg_volume_per_trade,
            'Unique_NFTs': trade_df['NFT_ID'].nunique()
        })

# Convert to DataFrame and summarize
comm_df = pd.DataFrame(comm_stats)
comm_df.to_csv('suspicious_communities.csv', index=False)

print(f"Detected {len(suspicious_communities)} suspicious communities")
print(f"Total suspicious volume: ${comm_df['Volume_USD'].sum():.2f}")
print(f"Total suspicious trades: {comm_df['Trade_Count'].sum()}")
print(f"Average community size: {comm_df['Address_Count'].mean():.2f} addresses")
print(f"Average trades per community: {comm_df['Trade_Count'].mean():.2f}")
print(f"Percentage of subset volume: {(comm_df['Volume_USD'].sum() / subset_multi_full['Price_USD'].sum()) * 100:.2f}%")
```

```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import time

start_time = time.time()

# Prepare features
features = subset_multi_full[['Price_USD', 'Datetime_updated_seconds']].copy()
features['Seller_Trade_Count'] = subset_multi_full.groupby('Seller_address')['Price_USD'].transform('count')
features['Buyer_Trade_Count'] = subset_multi_full.groupby('Buyer_address')['Price_USD'].transform('count')
features['Seller_Volume'] = subset_multi_full.groupby('Seller_address')['Price_USD'].transform('sum')
features['Price_Variance'] = subset_multi_full.groupby('NFT_ID')['Price_USD'].transform('var').fillna(0)

# Add trade frequency (trades per hour), optimized and finite
time_diffs = subset_multi_full.groupby('Seller_address')['Datetime_updated_seconds'].apply(
    lambda x: (x.max() - x.min()) / 3600 if len(x) > 1 else 0  # Use 0 for single trades
)
trade_freq = subset_multi_full.groupby('Seller_address')['Price_USD'].count() / time_diffs.replace(0, np.inf)  # Avoid div by 0
features['Seller_Trade_Freq'] = subset_multi_full['Seller_address'].map(trade_freq).fillna(0)  # 0 for no frequency

# Check for infinite or large values
print("Checking features for infinities or large values:")
print(features.describe())
print("Infinite values per column:\n", np.isinf(features).sum())

# Scale features (ensure finite)
features = features.replace([np.inf, -np.inf], 0)  # Replace any remaining inf with 0
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

# Fit Isolation Forest
iso_forest = IsolationForest(contamination=0.002, random_state=42)
iso_forest.fit(X_scaled)

# Predict anomalies
anomaly_scores = iso_forest.decision_function(X_scaled)
subset_multi_full['Anomaly_Score'] = anomaly_scores
subset_multi_full['Is_Anomaly'] = iso_forest.predict(X_scaled) == -1

# Quantify results
anomalies = subset_multi_full[subset_multi_full['Is_Anomaly']]
num_anomalies = len(anomalies)
anomaly_volume = anomalies['Price_USD'].sum()
subset_volume = subset_multi_full['Price_USD'].sum()
print(f"Detected anomalies: {num_anomalies} transactions")
print(f"Total anomalous volume: ${anomaly_volume:,.2f}")
print(f"Percentage of subset volume: {(anomaly_volume / subset_volume * 100):.2f}%")

# Overlap with cycles
wash_addresses = set()
for addr_list in cycles_df['Addresses']:
    wash_addresses.update(addr_list)
overlap = anomalies[
    anomalies['Seller_address'].isin(wash_addresses) | 
    anomalies['Buyer_address'].isin(wash_addresses)
]
overlap_volume = overlap['Price_USD'].sum()
print(f"Anomalies overlapping with cycle addresses: {len(overlap)} ({len(overlap)/num_anomalies*100:.1f}%)")
print(f"Overlap volume: ${overlap_volume:,.2f}")

# Top 10 anomalies
print("Top 10 anomalous transactions:")
print(anomalies.sort_values('Anomaly_Score').head(10)[
    ['Seller_address', 'Buyer_address', 'Price_USD', 'Anomaly_Score']
])

end_time = time.time()
print(f"Runtime: {end_time - start_time:.2f} seconds")
```
